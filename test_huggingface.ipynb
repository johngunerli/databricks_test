```
import pip
def import_or_install(package):
    try:
        __import__(package)
    except ImportError:
        pip.main(['install', package])   
```

```
from datasets import load_dataset
dataset = load_dataset("jxm/the_office_lines")
```

```
dataset["val"][0]
```

```
validation = dataset["val"].to_pandas()
train = dataset["train"].to_pandas()
test = dataset["test"].to_pandas()
#Contains a train split (47,927 lines), test split (5,991 lines) and validation split (5,991 lines).
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("MyApp").getOrCreate()

spark_val = spark.createDataFrame(validation)
spark_train = spark.createDataFrame(train)
spark_test = spark.createDataFrame(test)
```
